## Part A:Preprocessing and Initial Setup

```python
import numpy as np
import pandas as pd
from scipy.io import arff
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE, Isomap
from sklearn.metrics import pairwise_distances
import matplotlib.pyplot as plt
import seaborn as sns
```

So initially we have 103 features and 2417 data points, none of which contain any null values. So we're good to go

```python
# Convert each label row to a tuple, e.g. (0,1,0,...)
label_combos = [tuple(row) for row in y]
label_counts = pd.Series(label_combos).value_counts()

# Identify most common patterns
most_common_combo = label_counts.index[0]
multi_label_mask = np.array([tuple(row) == most_common_combo for row in y])

# Identify most frequent single-labels
single_label_counts = y.sum(axis=0)
most_common_labels = single_label_counts.argsort()[-2:][::-1]

# Create new target for coloring
color_target = np.full(y.shape[0], "Other", dtype=object)
color_target[y[:, most_common_labels[0]] == 1] = f"Label_{most_common_labels[0]}"
color_target[y[:, most_common_labels[1]] == 1] = f"Label_{most_common_labels[1]}"
color_target[multi_label_mask] = "MultiLabel_Combo"

pd.Series(color_target).value_counts()
```

```python
# Find most common multi-label combination
label_combos = [tuple(row) for row in y]
label_counts = pd.Series(label_combos).value_counts()

most_common_combo = label_counts.index[0]
most_common_count = label_counts.iloc[0]

# Convert the tuple (0/1 pattern) into readable class names
most_common_labels = [label_names[i] for i, val in enumerate(most_common_combo) if val == 1]

print("Most frequent multi-label combination:")
print(f" → Labels: {most_common_labels}")
print(f" → Occurs in {most_common_count} samples\n")

# Create mask for this combination
multi_label_mask = np.array([tuple(row) == most_common_combo for row in y])
```

```python
# Compute pairwise Euclidean distance matrix
dist_matrix = pairwise_distances(X_scaled, metric='euclidean')

plt.figure(figsize=(8,6))
sns.heatmap(dist_matrix[:100, :100], cmap='viridis')  # show first 100 for clarity
plt.title("Similarity Matrix (Euclidean Distances)")
plt.xlabel("Samples")
plt.ylabel("Samples")
plt.show()
```

## Part B : t-SNE and Veracity Inspection

```python
tsne = TSNE(n_components=2, perplexity=50, learning_rate='auto', init='pca', random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
for label in np.unique(color_target):
    mask = color_target == label
    plt.scatter(X_tsne[mask, 0], X_tsne[mask, 1], label=label, alpha=0.7)
plt.title("t-SNE Visualization of Yeast Data")
plt.legend()
plt.show()
```

Some clustering visible for Label 12, Label 11 seems infrequent. Also, t-SNE will tend to exaggerate even small separations. So one should take this with a pinch of salt when the other labels do not form as good of a cluster

Trying different perplexity values shows the same trend. At very high perplexities, other labels also start to form local clusters.

Perplexity represents the information entropy of the neighbors of each point. So higher perplexity means that farther neighbors are also included. This justifies the observation clearly.

In short,

- Low perplexity = narrow Gaussian = only a few near neighbors matter.
- High perplexity = wider Gaussian = more neighbors influence embedding.

- t-SNE preserves **local relationships**, revealing small clusters.
- Mixed colors inside clusters result in noisy/ambiguous labels.
- Isolated points are **outliers**.
- Samples from dense, mixed regions are hard to learn.


## Part C

```python
iso = Isomap(n_components=2, n_neighbors=50)
X_iso = iso.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
for label in np.unique(color_target):
    mask = color_target == label
    plt.scatter(X_iso[mask, 0], X_iso[mask, 1], label=label, alpha=0.7)
plt.title("Isomap Visualization of Yeast Data")
plt.legend()
plt.show()
```

What Isomap does is try to preserve the geodesic distances(distance measured over the surface of the manifold) when projecting the data into lower dimensions.

Local connections are defined using K Nearest neighbors. This generates a graph 

After checking through multiple values of \n_neighbors , the data does not show any behavioral problems. If this wasn't the case, however, it would indicate problems with the graph itself(disconnectedness, too dense, etc)

However, this is not the case here.

- Isomap preserves **global structure**, showing overall manifold shape.
- Compared to t-SNE, Isomap shows global curvature while t-SNE shows local separation better.
- Highly curved manifold => complex relationships => harder classification.


```python
fig, ax = plt.subplots(1, 2, figsize=(14,6))
for label in np.unique(color_target):
    mask = color_target == label
    ax[0].scatter(X_tsne[mask, 0], X_tsne[mask, 1], label=label, alpha=0.7)
    ax[1].scatter(X_iso[mask, 0], X_iso[mask, 1], label=label, alpha=0.7)

ax[0].set_title("t-SNE")
ax[1].set_title("Isomap")
plt.legend()
plt.show()
```

### Observations and Veracity Analysis
- **Noisy / Ambiguous Labels:** Overlapping colors in t-SNE plot.
- **Outliers:** Isolated points far from any cluster.
- **Hard-to-Learn Regions:** Mixed-color dense zones.
- **Global vs Local Structure:** t-SNE shows local clusters; Isomap shows global shape.
